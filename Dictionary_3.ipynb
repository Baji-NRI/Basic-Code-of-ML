```python
NRI500 = {
  "Roll no" : "NRI500",
  "Name" : "Sai Ram",
  "DOB" : "12/12/2012",
  "GPA" : 6.5,
  "isMarried" : False,
  "isStudent" : True,
}
```

```python
for _ in NRI500.keys():
    print(_)
```

```python
for _ in NRI500.values():
    print(_)
```

```python
for _ in NRI500.items():
    print(_)
```

```python
for key in NRI500:
    print(f"The key is {key} and value is {NRI500[key]}")
```

```python
for k,v in NRI500.items():
    print(f"The Key is {k} and value is{v}")
```
